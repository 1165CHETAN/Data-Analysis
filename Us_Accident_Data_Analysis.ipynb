## Percentage of missing values per columns

```python
missing_percentages=df.isna().sum().sort_values(ascending=False) / len(df)
missing_percentages
#this sorts missing values in percentages
```

```python
missing_percentages[missing_percentages !=0].plot(kind='barh')
#this sorts missing values in form of graphs
```

```python
cities=df.City.unique()
len(cities)
#total unique cities in us 
```

```python
cities_by_accidents = df.City.value_counts()
cities_by_accidents
#Top accidents in usa by cities
```

##Start Time

```python
df.columns
```

```python
df.Start_Time[0]
#Starting of accident
```

```python
df.Start_Time=pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
sns.distplot(df.Start_Time.dt.hour,bins=24,kde=False,norm_hist=True)
#Accidents are not happening at morning level in form of percentage
#High percentage is between 3PM to 6PM
#Second high percentage accident is b/w 6AM to 9AM morning
```

```python
sns.histplot(df.Start_Time.dt.hour,bins=24)
#Accidents are not happening at morning level in form of numbers
```

```python
sns.distplot(df.Start_Time.dt.dayofweek,bins=7,kde=False,norm_hist=True)
#Highest accident on friday
#Lowest at weekends
```

Is the distrubution of accidents by hour same on weekends as on weekdays?


```python
sundays_start_time=df.Start_Time[df.Start_Time.dt.dayofweek==6]
sns.distplot(sundays_start_time.dt.hour,bins=24,kde=False,norm_hist=True)
#After aftenoon more accidents happens
```

```python
monday_start_time=df.Start_Time[df.Start_Time.dt.dayofweek==0]
sns.distplot(monday_start_time.dt.hour,bins=24,kde=False,norm_hist=True)
#After aftenoon more accidents happens
#Between people going to work and returning from office
```
