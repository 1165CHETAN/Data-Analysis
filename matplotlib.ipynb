**mat2**

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
#M1
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max",c='red',linewidth='2')
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend()
```

```python
#M2
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max",c='red')
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='lower left',shadow=True)
```

**mat3(Bargraphs)**

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
x = np.arange(len(company))
x
```

```python
plt.bar(x,revenue,align='center',color='green',label='revenue')
plt.xticks(x,company)
plt.legend(loc='upper right')
```

```python
profit=[40,2,34,12]
```

```python
plt.bar(x-0.2,revenue, width=0.4, label="Revenue")
plt.bar(x+0.2,profit, width=0.4,label="Profit")

plt.xticks(x,company)
```

```python
plt.barh(x-0.2,revenue, label="Revenue",color='violet')
plt.barh(x+0.2,profit,label="Profit")

plt.yticks(x,company)
plt.legend()
```

**mat4(Histograms)**

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

```python


men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([men,women], bins=[80,100,125,150], rwidth=0.65, color=['pink','skyblue'],label=['men','women'])
plt.legend()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=[80,100,125,150],color='orange', orientation='horizontal',rwidth=0.85)
```

**mat5(Piechart)**

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
val = [1400,600,300,410,250]
label = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(val,labels=label,shadow=True,explode=[0.1,0,0,0,0],startangle=90,autopct='%1.1f%%')
plt.savefig("piechart.png")
```
