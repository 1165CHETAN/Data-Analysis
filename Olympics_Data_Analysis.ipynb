```python
athlete=pd.read_csv('./120-years-of-olympic-history-athletes-and-results/athlete_events.csv')
regions=pd.read_csv('./120-years-of-olympic-history-athletes-and-results/noc_regions.csv')
```

```python
regions.head()
```

```python
#join the dataframes
#regions has been added to athletes datasets
athletes_df= athlete.merge(regions,how='left',on='NOC')
athletes_df.head()

```

```python
#top countries participating in olympics from 1896
top_10_countries = athletes_df. Team.value_counts().sort_values(ascending=False).head(10) 
top_10_countries
```

```python
#top countries participating in olympics from 1896 using bargraph
plt.figure(figsize=(12,6))
plt.title('Overall Prticipation By Countries')
sns.barplot(x=top_10_countries.index,y=top_10_countries,palette='Set2')
```

```python
plt.figure(figsize=(12, 6))

plt.title("Age distribution of the athletes")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist (athletes_df.Age, bins=np.arange(10,80,2),color='orange', edgecolor= 'white')
```

```python
gender_counts=athletes_df.Sex.value_counts()
gender_counts
#to calulate male and female
```

```python
plt.figure(figsize=(12,6))
plt.title('Gebder Distrubution')
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=150,shadow=True);
#to calulate male and female using pie chart
#autopct shows value in percentage using string formatting
```

```python
#Total medals
athletes_df.Medal.value_counts()
```

```python
female_participants=athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=='Summer')][['Sex','Year']]
female_participants=female_participants.groupby('Year').count().reset_index()
female_participants.head()
female_participants.tail()
```

```python
WomenOlympics=athletes_df[(athletes_df.Sex=='F')&(athletes_df.Season=='Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=WomenOlympics,palette='Spectral')
plt.title('Women participation')
#women participation increses year by year
```

```python
part=WomenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes Over Time')
#Line graph since 1900 to 2020
#Decrease in 1980
```

```python
gold_medals=athletes_df[(athletes_df.Medal=='Gold')]
gold_medals.head()
#top 5 gold medalist
```

```python
#Gold medal beyond 60 are 6
gold_medals=gold_medals[np.isfinite(gold_medals['Age'])]
```

```python
gold_medals['ID'][gold_medals['Age']>60].count()
```

```python
sporting_event=gold_medals['Sport'][gold_medals['Age']>60]
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Gold medals for athletes over 60 years old')
```

```python
#Total gold medals by country
gold_medals.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalgoldmedals=gold_medals.Region.value_counts().reset_index(name='Medal').head(6)
g=sns.catplot(x='index',y='Medal',data=totalgoldmedals,
              height=5,kind='bar',palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 Countries')
g.set_ylabels('Number of Medals')
plt.title('Gold Medals per country')
```
